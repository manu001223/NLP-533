```python
import PyPDF2
import docx
import pandas as pd

# ---------- 1. Read TXT file ----------
def read_txt(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        return f.read()

# ---------- 2. Read PDF file ----------
def read_pdf(file_path):
    text = ""
    with open(file_path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            text += page.extract_text()
    return text

# ---------- 3. Read DOCX file ----------
def read_docx(file_path):
    doc = docx.Document(file_path)
    text = ""
    for para in doc.paragraphs:
        text += para.text + "\n"
    return text

# ---------- 4. Read CSV file ----------
def read_csv(file_path):
    df = pd.read_csv(file_path)
    return df.to_string()

# ---------- Provide file paths ----------
txt_file = "/content/Data_Dictionary.txt"
pdf_file = "/content/Epstein_Case_Timeline.pdf"
docx_file = "/content/School of Advanced Computing (1).docx"
csv_file = "/content/Cleaned_Bookings_PowerBI.csv"

# ---------- Extract text ----------
txt_text = read_txt(txt_file)
pdf_text = read_pdf(pdf_file)
docx_text = read_docx(docx_file)
csv_text = read_csv(csv_file)

# ---------- Combine all ----------
combined_text = txt_text + "\n" + pdf_text + "\n" + docx_text + "\n" + csv_text

# ---------- Save unified text ----------
with open("/content/final_output.txt", "w", encoding="utf-8") as f:
    f.write(combined_text)

print("All file formats successfully converted into plain text.")
```
