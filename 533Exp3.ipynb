```python
Pattern = r'\s'
data['char_count'] = data['twitts'].apply(lambda x: len(re.sub(Pattern, '', x)))
```

```python
data
```

```python
data['Word_count'] = data['twitts'].apply(lambda x: len(x.split()))
```

```python
data
```

```python
data['avg_word_legnth'] = data['char_count']/data['Word_count']
```

```python
data
```

```python
pattern1 = r'\w+@\w+.\w+'
```

```python
data['email'] = data['twitts'].apply(lambda x: re.findall(pattern1, x))
display(data['email'].sample())
```

```python
pattern=r'\b[A-Za-z._+-]+@[A-Za-z.][A-Z|a-z]{2,}\b'
data['email']=data['twitts'].apply(lambda x: re.findall(pattern=pattern,string=x))
data['email'].sample(5)
```
